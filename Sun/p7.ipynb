```python
abcmileage = []
for i in range(0,len(c),1100):
    abcmileage.append(c[i:i+1100])
abcmileage
# This is the previous version of the matrix mileage; 
```

```python

matrix = abcmileage
def change(matrix):
    m = len(matrix)
    n = len(matrix[0])
    result = [[] for i in range(n)]
    for i in range(m):
        for j in range(n):
            if i == m:
                result[j].append(0)
            else:
                result[j].append(matrix[i][j])
    return result

mileage = change(matrix)
print(mileage)
# m*n becomes n*m
```

```python
abctrailers = [a,a,a,a,a,a,a,a,a,a]
```

```python

matrix = abctrailers
def change(matrix):
    m = len(matrix)
    n = len(matrix[0])
    result = [[] for i in range(n)]
    for i in range(m):
        for j in range(n):
            if i == m:
                result[j].append(0)
            else:
                result[j].append(matrix[i][j])
    return result

trailers = change(matrix)
print(trailers)
```

```python
m = Model ('Optimization Model')
#m.remove(m.getVars())
```

```python
fixcost = 200
varcost = 0.75
dccapacity = 12000
```

```python
x= np.array([[m.addVar(vtype=GRB.BINARY, name='x'+str(i)+'_'+str(j)) for j in range(len(cc))]for i in range(len(rr))])
m.update()
```

```python
for i in range(len(a)):
    m.addConstr(sum(x[i,])*a[i],GRB.EQUAL,a[i])
for j in range(10):
    m.addConstr(quicksum(x[:,j][i]*a[i] for i in range(1100)),GRB.LESS_EQUAL,dccapacity)
m.update()
```

```python
m.setObjective(quicksum(trailers[i][j]*mileage[i][j]*x[i,j] for j in range(10) for i in range(1100))*varcost + (quicksum(trailers[i][j]*x[i,j] for j in range(10) for i in range(1100)))*fixcost,GRB.MINIMIZE)
m.update()
```

```python
m.optimize()
```
